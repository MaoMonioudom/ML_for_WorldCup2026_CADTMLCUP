# Notebook 02 — Retrain with New Features
New features added:
- H2H last 5 matches
- Recent form last 5 matches
- Competition type (encoded 1–4)
- Home win rate vs Away win rate separately
- FIFA ranking

```python
# ============================================================
# CELL 1 — Process FIFA Ranking: keep latest rank per team
# ============================================================
import pandas as pd
import os

ranking = pd.read_csv('../data/fifa_ranking-2024-06-20.csv')
ranking['rank_date'] = pd.to_datetime(ranking['rank_date'])

print('Raw ranking shape:', ranking.shape)
print(ranking.head())
```

```python
# ============================================================
# CELL 2 — Keep only the latest ranking per team & save
# ============================================================

# Sort by date and keep the last (most recent) entry per team
latest_ranking = (
    ranking.sort_values('rank_date')
           .groupby('country_full', as_index=False)
           .last()
)[['country_full', 'rank', 'total_points', 'rank_date']]

latest_ranking.columns = ['team', 'fifa_rank', 'fifa_points', 'rank_date']

os.makedirs('../dataII', exist_ok=True)
latest_ranking.to_csv('../dataII/rankings.csv', index=False)

print('Saved rankings.csv — shape:', latest_ranking.shape)
print(latest_ranking.head(10))
```
